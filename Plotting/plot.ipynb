```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('LiH.csv')
```

```python
# Plot Bond length vs CN_A and CN_C
plt.figure(figsize=(10, 6))

# Plot Bond length vs CN_A
plt.plot(data['Bond Distance'], data['CN_A'], label='CN of H', marker='o')

# Plot Bond length vs CN_C
plt.plot( data['Bond Distance'],data['CN_C'], label='CN of Li', marker='x')

# Add labels, title, and legend
plt.ylabel('Coordination Number (CN)')
plt.xlabel('Bond Length (Å)')
plt.title('Bond Length vs CN for LiH')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
