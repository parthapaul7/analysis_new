```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create combined x-axis label

full_li = merged_df['Zn Sites']
merged_df['Anion Site'] = (merged_df['Counter Ion Sites'] / full_li).round(2)

merged_df.sort_values(by='Anion Site',inplace=True, ascending=True)
merged_df['Structure_SG'] = "Zn-" + symbol+(merged_df['Counter Ion Sites'] / full_li).round(2).astype(str) 
# Set data
x_labels = merged_df['Structure_SG']
values = merged_df['Result_1']

# Create figure
plt.figure(figsize=(10, 6))

# Create bar plot
x = np.arange(len(x_labels))  # numeric x-axis positions
plt.bar(x, values, width=0.6)

# X-axis setup
plt.xticks(x, x_labels, rotation=60, ha='right')
plt.xlabel('Structure')
plt.ylabel('Energy (mEh)')
# plt.title('Grouped Bar Plot of Result_1 by Structure and SpaceGroup')
# plt.legend(title='Metric')
plt.tight_layout()

# Save or show
# plt.savefig(f'chemical_space/Zn_{symbol}_compounds.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 1. Build a display label with subscripted CN
def make_disp_label(sg):
    # sg is like "Li-N0.33" or "K-N1.0"
    metal, cn_str = sg.split(f'-{element}')  # ["Li", "0.33"]
    cn_str = f"{float(cn_str):.2f}"  # Ensure two decimal places
    return rf"{cn_str}"

merged_df['DispLabel'] = merged_df['Structure_SG'].apply(make_disp_label)

# 2. Plot
x = np.arange(len(merged_df))
values = merged_df['Result_1']
labels = merged_df['DispLabel']

plt.figure(figsize=(16, 12))
plt.rcParams['font.family'] = 'Arial'
plt.bar(x, values, width=0.6, color='teal')

# 3. Use the LaTeX labels and set y-axis limit
plt.xticks(x, labels, rotation=45, ha='right', fontsize=38)
plt.yticks(fontsize=38)
plt.ylim(0, 0.4)  # Set y-axis limit to 0.5
plt.ylabel('Energy (eV)', fontsize=38)
plt.xlabel(rf'$\mathrm{{Zn{symbol}_X}}$', fontsize=38)

plt.tight_layout()
# plt.savefig(f'chemical_space/Cu_{symbol}_compounds.png', dpi=300, bbox_inches='tight')
plt.show()
```
